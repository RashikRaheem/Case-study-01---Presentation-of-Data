```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### 1. Read the dataset to the python environment.

```python
df=pd.read_csv('cars_data.csv')
df
```

### 2. Check for the null values present in the dataset.

```python
df.isnull().sum().sum()   #Zero null values
```

### 3. Plot a bar graph of male vs female buyers participated in the sales. 

```python
Male_Count=df['Buyer Gender'].isin(['Male']).sum()
Male_Count
```

```python
Female_Count=df['Buyer Gender'].isin(['Female']).sum()
Female_Count
```

```python
x=['Male','Female']
y=[Male_Count,Female_Count]
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]//2, y[i], ha = 'center',color='w',size='large')

plt.bar(x,y,color=['blue','y']);
plt.title("Male vs Female buyers")
addlabels(x,y)
plt.ylabel="Number of buyers"
plt.xlabel=""
plt.show()
```

### 4. Find the top 5 cars based on their sales price.

```python
Top=df.sort_values(by=['Sale Price'],ascending=False).head()
```

```python
Top[['Make','Model','Sale Price']].reset_index()  #Top 5 Cars
```
